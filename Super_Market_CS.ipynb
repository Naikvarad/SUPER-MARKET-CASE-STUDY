```python
def outlier_treatment(datacolumn):
 sorted(datacolumn)
 Q1,Q3 = np.percentile(datacolumn,[25,75])
 IQR = Q3-Q1
 lower_range = Q1 - (1.5*IQR)
 upper_range = Q3 + (1.5*IQR)
 return lower_range,upper_range
```

# What kind of customers responded the most in the last campaign?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the data to include only rows with responses
response_data = d[d['Response'] == 1]

# Group by Education and Marital_Status, and calculate the response count
response_by_segment = response_data.groupby(['Education', 'Marital_Status']).size().reset_index(name='Response Count')

# Check if there are any responses
if not response_by_segment.empty:
    # Sort the results by response count in descending order
    response_by_segment = response_by_segment.sort_values(by='Response Count', ascending=False)

    # Plot the response count by segment
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Response Count', y='Education', hue='Marital_Status', data=response_by_segment, orient='h')
    plt.title('Response Count by Education and Marital Status')
    plt.xlabel('Response Count')
    plt.ylabel('Education Level')
    plt.show()
else:
    print("No responses in the specified Education and Marital_Status segments.")
```

# Did certain customers prefer Web Purchase over Store Purchases? If yes, how were their
 profiles different?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select relevant columns for analysis
web_purchase_column = 'NumWebPurchases'
store_purchase_column = 'NumStorePurchases'
categorical_columns = ['Education', 'Marital_Status']

# Melt the dataframe to a long format for easier plotting
df_melted = pd.melt(d, id_vars=categorical_columns, value_vars=[web_purchase_column, store_purchase_column], 
                    var_name='Purchase Type', value_name='Number of Purchases')

# Create a bar plot for average number of web and store purchases by education and marital status
plt.figure(figsize=(12, 8))
sns.barplot(x='Number of Purchases', y='Education', hue='Marital_Status', data=df_melted, orient='h', 
            hue_order=d['Marital_Status'].unique())
plt.title('Average Number of Web and Store Purchases by Education and Marital Status')
plt.xlabel('Average Number of Purchases')
plt.ylabel('Education Level')
plt.show()
```

# Is customer complaints an area of concern that should be closely looked at? Show
 appropriate metrics to support your response

```python
complaint_rate = (d['Complain'].sum() / len(d)) * 100
print(f'Complaint Rate: {complaint_rate:.2f}%')
```

```python
product_complaints = d.groupby('Complain')[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum()
print('Complaints by Product Category:')
print(product_complaints)
```

```python
country_complaints = d.groupby('Country')['Complain'].sum()
print('Complaints by Country:')
print(country_complaints)
```

```python
churn_rate_with_complaints = d[d['Complain'] == 1]['Response'].mean()
churn_rate_without_complaints = d[d['Complain'] == 0]['Response'].mean()
print(f'Churn Rate with Complaints: {churn_rate_with_complaints:.2%}')
print(f'Churn Rate without Complaints: {churn_rate_without_complaints:.2%}')
```
